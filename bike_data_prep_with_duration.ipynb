```python
import pandas as pd
import datetime as dt
```

```python
bike_may_df = pd.read_csv("Resources/JC-202305-citibike-tripdata.csv")
bike_may_df.head()
```

```python
bike_may_df.info()
```

```python
bike_april_df = pd.read_csv("Resources/JC-202304-citibike-tripdata.csv")
bike_april_df.head()
```

```python
bike_april_df.info()
```

```python
bike_march_df = pd.read_csv("Resources/JC-202303-citibike-tripdata.csv")
bike_march_df.head()
```

```python
bike_march_df.info()
```

```python
all_bikes_df = pd.concat([bike_may_df, bike_april_df, bike_march_df], ignore_index=True)
```

```python
all_bikes_df.head()
```

```python
all_bikes_df.info()
```

```python
all_bikes_df = all_bikes_df.dropna()
```

```python
all_bikes_df.info()
```

```python
all_bikes_df.head()
```

```python
all_bikes_df['started_at'] = pd.to_datetime(all_bikes_df['started_at'])
all_bikes_df['ended_at'] = pd.to_datetime(all_bikes_df['ended_at'])
```

```python
all_bikes_df.head()
```

```python
all_bikes_df.info()
```

```python
all_bikes_df['ride_duration_mins'] = all_bikes_df['ended_at'] - all_bikes_df['started_at']
```

```python
all_bikes_df.head()
```

```python
all_bikes_df['ride_duration_mins'] = all_bikes_df['ride_duration_mins'].dt.total_seconds() / 60
```

```python
all_bikes_df.head()
```

```python
all_bikes_df.to_csv("Resources/all_bikes_spring2023_withduration.csv")
```
